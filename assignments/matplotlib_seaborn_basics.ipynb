# **Matplotlib**



## **1. What is Matplotlib?**

Matplotlib is a Python library for creating charts and graphs.

**When to use it:**
- Visualizing data from lists, NumPy arrays, or files
- Creating reports with charts
- Exploring data to understand patterns
- Presenting your findings

**Think of it as:** A digital canvas where you can draw lines, bars, points, and more!

**Import the library:**

````python
import matplotlib.pyplot as plt

# This tells Jupyter to show plots in the notebook
%matplotlib inline
````

## **3. Your First Plot: A Simple Line Graph**

Let's plot how your study hours increased over 5 days:

````python
# Data: days and hours studied
days = [1, 2, 3, 4, 5]
hours = [2, 3, 4, 3, 5]

# Create the plot
plt.plot(days, hours)

# Display the plot
plt.show()
````

## **4. Making Your Plot Look Professional**

A good plot needs labels and a title so people understand what they're looking at!

### **Adding Title and Labels**

````python
days = [1, 2, 3, 4, 5]
hours = [2, 3, 4, 3, 5]

plt.plot(days, hours)

# Add title and labels
plt.title('My Study Hours This Week')
plt.xlabel('Day')
plt.ylabel('Hours Studied')

plt.show()
````

### **Adding a Grid for Better Readability**

````python
days = [1, 2, 3, 4, 5]
hours = [2, 3, 4, 3, 5]

plt.plot(days, hours)
plt.title('My Study Hours This Week')
plt.xlabel('Day')
plt.ylabel('Hours Studied')

# Add grid lines
plt.grid(True)

plt.show()
````

### **Customizing Line Color and Style**

````python
days = [1, 2, 3, 4, 5]
hours = [2, 3, 4, 3, 5]

# Color: green, marker: circle (o), line style: dashed (--)
plt.plot(days, hours, color='green', marker='o', linestyle='--')

plt.title('My Study Hours This Week')
plt.xlabel('Day')
plt.ylabel('Hours Studied')
plt.grid(True)

plt.show()
````

**Common colors:** `'red'`, `'blue'`, `'green'`, `'purple'`, `'orange'`

**Common markers:** `'o'` (circle), `'s'` (square), `'^'` (triangle), `'*'` (star)

**Line styles:** `'-'` (solid), `'--'` (dashed), `'-.'` (dash-dot), `':'` (dotted)

## **5. Common Plot Types**

Different data needs different visualizations. Let's explore the most useful ones!

### **Line Plot** (We just learned this!)

**Best for:** Showing trends over time (temperature, stock prices, progress)

````python
# Temperature throughout the day
hours = [6, 9, 12, 15, 18, 21]
temperature = [15, 18, 25, 28, 24, 19]

plt.plot(hours, temperature, color='red', marker='o')
plt.title('Temperature Throughout the Day')
plt.xlabel('Hour')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
````

### **Bar Chart**

**Best for:** Comparing categories (sales by product, scores by student)

````python
# Sales by product
products = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sales = [45, 60, 30, 50]

plt.bar(products, sales, color='skyblue')
plt.title('Fruit Sales This Week')
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.show()
````

**Horizontal bar chart:** Use `plt.barh()` instead of `plt.bar()`

````python
products = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sales = [45, 60, 30, 50]

plt.barh(products, sales, color='lightgreen')
plt.title('Fruit Sales This Week')
plt.xlabel('Units Sold')
plt.ylabel('Product')
plt.show()
````

### **Scatter Plot**

**Best for:** Showing relationships between two variables (height vs weight, study time vs grades)

````python
# Study hours vs Exam scores
study_hours = [2, 3, 4, 5, 6, 7, 8]
exam_scores = [55, 60, 68, 75, 82, 88, 95]

plt.scatter(study_hours, exam_scores, color='purple', s=100)
plt.title('Study Hours vs Exam Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
````

**Note:** `s=100` controls the size of the dots. Try changing it!

### **Histogram**

**Best for:** Showing distribution of data (test scores, ages, heights)

````python
# Test scores of 20 students
scores = [55, 60, 62, 65, 68, 70, 72, 75, 75, 78,
          80, 82, 85, 85, 88, 90, 92, 95, 98, 100]

plt.hist(scores, bins=5, color='orange', edgecolor='black')
plt.title('Distribution of Test Scores')
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.show()
````

**Note:** `bins=5` divides data into 5 groups. More bins = more detail!

## **Adding Legends to Your Plots**

Legends help identify different data series in your plots. They're essential when you have multiple lines or datasets!

### **Legend with Bar Charts**

**Legend positions:**
- `'upper left'`, `'upper right'`, `'lower left'`, `'lower right'`
- `'center'`, `'upper center'`, `'lower center'`
- `'center left'`, `'center right'`
- `'best'` (default - matplotlib chooses the best spot)

````python
import numpy as np

# Sales data for two quarters
products = ['Product A', 'Product B', 'Product C', 'Product D']
q1_sales = [45, 30, 55, 40]
q2_sales = [50, 35, 60, 45]

x = np.arange(len(products))  # Position for bars
width = 0.35  # Width of each bar

# Create bars with labels
plt.bar(x - width/2, q1_sales, width, label='Q1', color='skyblue')
plt.bar(x + width/2, q2_sales, width, label='Q2', color='lightcoral')

plt.title('Quarterly Sales Comparison')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.xticks(x, products)  # Set product names as x-axis labels
plt.legend(loc='best')

plt.show()
````

**Important:**
- Add `label='Your Label'` to each plot
- Call `plt.legend()` to display the legend

### **Pie Chart**

**Best for:** Showing parts of a whole (percentages, proportions)

````python
# Time spent on daily activities
activities = ['Sleep', 'Work', 'Exercise', 'Leisure']
hours = [8, 8, 2, 6]

plt.pie(hours, labels=activities, autopct='%1.1f%%')
plt.title('Daily Time Distribution')
# plt.legend()
plt.show()
````

**Note:** `autopct='%1.1f%%'` shows percentages on the slices (1 decimal place)

**Pie chart options:**
- `explode` = pull out specific slices (0.1 means 10% of the pie radius)
- `shadow=True` = adds a shadow effect
- `startangle=90` = rotate the starting position
- `colors` = custom colors for each slice

````python
# Monthly budget breakdown
categories = ['Rent', 'Food', 'Transport', 'Entertainment', 'Savings']
amounts = [1200, 400, 200, 150, 250]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc']

# Explode the first slice (Rent) to highlight it
explode = (0.1, 0, 0, 0, 0)

plt.pie(amounts, labels=categories, colors=colors, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Monthly Budget Breakdown')
plt.show()
````

## **6. Figure Size and Saving Plots**

### **Changing Figure Size**

````python
# Create a larger plot
plt.figure(figsize=(10, 6))  # width=10 inches, height=6 inches

days = [1, 2, 3, 4, 5]
hours = [2, 3, 4, 3, 5]

plt.plot(days, hours, color='blue', marker='s', linewidth=2)
plt.title('My Study Hours - Large Version', fontsize=16)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Hours', fontsize=12)
plt.grid(True)
plt.show()
````

### **Saving Your Plot to a File**

````python
days = [1, 2, 3, 4, 5]
hours = [2, 3, 4, 3, 5]

plt.plot(days, hours, color='red', marker='o')
plt.title('My Study Hours This Week')
plt.xlabel('Day')
plt.ylabel('Hours Studied')
plt.grid(True)

# Save before showing
plt.savefig('my_study_plot.png', dpi=300, bbox_inches='tight')
plt.show()
````

**Options:**
- `dpi=300` = high quality (for printing)
- `bbox_inches='tight'` = no extra white space
- You can save as: `.png`, `.jpg`, `.pdf`, `.svg`

# **Seaborn Tutorial**

## **1. What is Seaborn?**

Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive statistical graphics.

**Why use Seaborn instead of Matplotlib?**
- Beautiful default styles and color palettes
- Built-in statistical plots
- Less code for complex visualizations
- Great for data analysis and exploration

**Think of it as:** Matplotlib with a professional design upgrade!

## **2. Installation and Import**

If you don't have Seaborn installed, run:
```bash
pip install seaborn
```

Seaborn usually comes pre-installed with Anaconda and Google Colab!

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline

print("Seaborn version:", sns.__version__)
````

## **3. Built-in Datasets**

Seaborn comes with several datasets perfect for practice!

````python
# Load the famous tips dataset
tips = sns.load_dataset('tips')

# Show first few rows
print(tips.head())
print("\nDataset shape:", tips.shape)
print("\nColumn info:")
print(tips.info())
````

## **4. Your First Seaborn Plot**

Let's create a simple scatter plot!

````python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('Bill Amount vs Tip')
plt.show()
````

## **5. Essential Seaborn Plots**

### **Scatter Plot with Colors (hue)**

Add a third dimension using color!

````python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
plt.title('Bill vs Tip (colored by time)')
plt.show()
````

### **Line Plot**

````python
data = pd.DataFrame({
    'day': range(1, 8),
    'sales': [120, 150, 140, 180, 200, 190, 220]
})

sns.lineplot(data=data, x='day', y='sales', marker='o')
plt.title('Weekly Sales Trend')
plt.xlabel('Day')
plt.ylabel('Sales ($)')
plt.show()
````

### **Bar Plot**

````python
sns.barplot(data=tips, x='day', y='tip', errorbar=None)
plt.title('Average Tip by Day of Week')
plt.ylabel('Average Tip ($)')
plt.show()
````

### **Count Plot**

````python
sns.countplot(data=tips, x='day')
plt.title('Number of Customers per Day')
plt.ylabel('Count')
plt.show()
````

### **Histogram**

````python
sns.histplot(data=tips, x='total_bill', bins=20)
plt.title('Distribution of Bill Amounts')
plt.xlabel('Total Bill ($)')
plt.show()
````

### **Box Plot**

````python
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('Bill Distribution by Day')
plt.ylabel('Total Bill ($)')
plt.show()
````

### **Violin Plot**

````python
sns.violinplot(data=tips, x='time', y='total_bill')
plt.title('Bill Distribution: Lunch vs Dinner')
plt.ylabel('Total Bill ($)')
plt.show()
````

### **Heatmap**

````python
correlation = tips[['total_bill', 'tip', 'size']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
````

## **6. Customization**

### **Color Palettes**

palettes = ['deep', 'muted', 'pastel', 'bright', 'dark', 'colorblind', 'Set1', 'Set2', 'Set3', 'Paired', 'tab10']

````python
sns.barplot(data=tips, x='day', y='tip', palette='Set2', hue='day', legend=False)
plt.title('Average Tip by Day (Set2 palette)')
plt.ylabel('Average Tip ($)')
plt.show()
````

### **Styles**

````python
sns.set_style('darkgrid')
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
plt.title('Scatter Plot with Dark Grid Style')
plt.show()

sns.set_style('whitegrid')  # Reset to default
````

## **7. Advanced: Pair Plot**

````python
sns.pairplot(tips, hue='time')
plt.show()
````
