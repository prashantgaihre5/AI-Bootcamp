# Library Management System



```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def __str__(self):
        return f"'{self.title}' by {self.author}"
```

```python
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print(f"Added: {book}")

    def remove_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                self.books.remove(book)
                print(f"Removed: {book}")
                return
        print(f"Book with title '{title}' not found.")

    def search_book(self, query):
        found_books = [
            book for book in self.books
            if query.lower() in book.title.lower() or query.lower() in book.author.lower()
        ]

        if found_books:
            print(f"\nSearch Results for '{query}':")
            for book in found_books:
                print(f" {book}")
        else:
            print(f"\nNo books found matching '{query}'.")

    def display_books(self):
        if not self.books:
            print("\nThe library is currently empty.")
        else:
            print("\nLibrary Collection:")
            for book in self.books:
                print(f" {book}")


```

```python
def main():
    my_library = Library()

    my_library.add_book(Book("b1", "aaa"))
    my_library.add_book(Book("b2", "bbb"))
    my_library.add_book(Book("b3", "ccc"))

    my_library.display_books()

    my_library.search_book("b1")
    my_library.search_book("b2")

    my_library.remove_book("b2")

    my_library.display_books()
```

```python
main()
```
